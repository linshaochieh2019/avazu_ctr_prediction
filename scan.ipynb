scan train dataset and find uid_click_counter
- how many times each uid appeared?
- user clicking history in an 0101 format?

```python
train_path = '/Users/linshoahchieh/Documents/Kaggle/CTR_prediction/data/train.gz'
save_path = '/Users/linshoahchieh/Documents/Kaggle/CTR_prediction/helpers/uid_click_counter.pkl'
        
chunksize = 1000000
uid_click_counter = Counter()
stopper = 0
for c in pd.read_csv(train_path, chunksize=chunksize):
    train = c
    # Get uid and then count
    train['uid'] = train['device_ip'] + train['device_model']
    uid_click_counter += Counter(train[train['click'] == 1]['uid'])
    stopper += 1
    if stopper % 5 == 4:
        with open(save_path, 'wb') as f: #save counter once in a while
            pickle.dump(uid_click_counter, f)
        print('Iteration index: ', stopper)
    break
```
